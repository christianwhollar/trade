```python
# Question 1: Get Daily, Weekly Volatility from Annual Volatility
# Assume Contract Price == Forward Price
def get_daily_and_weekly_vol(contract_price: float, vols: list) -> tuple[list[float], list[float]]:
    daily_vols, weekly_vols = [], []

    for vol in vols:
        
        n = 1 # number of standard deviations
        sigma = vol/100 # annual volatility or standard deviation
        F = contract_price # forward price
        
        t = 1/256 # time period
        daily_vols.append(n * F * sigma * np.sqrt(t))

        t = 1/52 # time period
        weekly_vols.append(n * F * sigma * np.sqrt(t))

    return daily_vols, weekly_vols

# Question 1.a
contract_price = 78
vols = [20, 30, 40, 50]
daily_vols, weekly_vols = get_daily_and_weekly_vol(contract_price, vols)
print(daily_vols)
print(weekly_vols)

# Question 1.b
contract_price = 13250
vols = [10, 15, 20, 25]
daily_vols, weekly_vols = get_daily_and_weekly_vol(contract_price, vols)
print(daily_vols)
print(weekly_vols)

# Question 1.c
contract_price = 1.6270
vols = [8, 10, 12, 14]
daily_vols, weekly_vols = get_daily_and_weekly_vol(contract_price, vols)
print(daily_vols)
print(weekly_vols)

# Question 1.d
contract_price = 669
vols = [15, 23, 31, 30]
daily_vols, weekly_vols = get_daily_and_weekly_vol(contract_price, vols)
print(daily_vols)
print(weekly_vols)

# Question 1.e
contract_price = 3187
vols = [13, 17, 21, 25]
daily_vols, weekly_vols = get_daily_and_weekly_vol(contract_price, vols)
print(daily_vols)
print(weekly_vols)
```

```python
# Question 2
def get_price_standard_deviations(futures_price: float, vol: float, t: float, t_unit: str) -> float:
    sigma = vol/100 # annual volatility or standard deviation
    F = futures_price
    t_convert = t/256 if t_unit == "d" else t/52 if t_unit == "w" else t/12 if t_unit == "m" else 0
    std_dev = F * sigma * np.sqrt(t_convert)

    return [
        F - std_dev,
        F + std_dev,
        F - 2 * std_dev,
        F + 2 * std_dev
    ]

# Question 2.a
print(
    get_price_standard_deviations(
        futures_price=226,
        vol=21,
        t=14, 
        t_unit="w",
    )
)

# Question 2.b
print(
    get_price_standard_deviations(
        futures_price=1869,
        vol=14,
        t=11, 
        t_unit="m",
    )
)

# Question 2.c
print(
    get_price_standard_deviations(
        futures_price=103.82,
        vol=9.5,
        t=116, 
        t_unit="d",
    )
)

# Question 2.d
print(
    get_price_standard_deviations(
        futures_price=16.97,
        vol=38,
        t=5, 
        t_unit="m",
    )
)

# Question 2.e
print(
    get_price_standard_deviations(
        futures_price=9.623,
        vol=18.25,
        t=23, 
        t_unit="w",
    )
)
```
